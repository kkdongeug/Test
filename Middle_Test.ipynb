```python
# 1. YOLO 설치
%pip install ultralytics --upgrade -q

# 2. 이미지 업로드
from google.colab import files
uploaded = files.upload()
image_path = next(iter(uploaded))

# 3. YOLOv8 모델 로드 및 이미지 분석
from ultralytics import YOLO
model = YOLO('yolov8x.pt')
results = model(image_path)
result = results[0]

# 4. 객체 수 세기
from collections import Counter
names = model.names
class_ids = result.boxes.cls.tolist()
class_names = [names[int(cls_id)] for cls_id in class_ids]
object_counts = Counter(class_names)
total_objects = len(class_names)

# 5. 이미지 시각화
import cv2
from PIL import Image

plot_img = result.plot()
plot_img = cv2.cvtColor(plot_img, cv2.COLOR_BGR2RGB)
img = Image.fromarray(plot_img)

# 6. 결과 출력 및 다운로드
import IPython.display as display
output_path = 'yolo_result.jpg'
img.save(output_path)
display.display(img)
files.download(output_path)
```
